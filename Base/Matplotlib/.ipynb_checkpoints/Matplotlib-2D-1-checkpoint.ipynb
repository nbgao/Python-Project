# 1. 2D图像绘制

## 1.1 线形图 plot

```python
import matplotlib.pyplot as plt
import numpy as np

X = np.linspace(-2*np.pi, 2*np.pi, 1000)
y = np.sin(X)

plt.plot(X,y)
plt.show()
```

## 1.2 柱形图 bar

```python
X = np.linspace(-2*np.pi, 2*np.pi, 15)
y = np.cos(X)
plt.bar(X,y)
plt.show()
```

## 1.3 散点图 scatter

```python
X = np.random.normal(0,1,1000)
y = np.random.normal(0,1,1000)

plt.figure(figsize=(6,6))
plt.scatter(X,y)
plt.show()
```

## 1.4 饼状图 pie

```python
Z = [1,2,3,4,5]

plt.figure(figsize=(5,5))
plt.pie(Z)
plt.show()
```

## 1.5 量场图 quiver

```python
X, y =np.mgrid[0:10, 0:10]

plt.figure(figsize=(6,6))
plt.quiver(X, y)
plt.show()
```

## 1.6 等高线图 contourf

```python
x = np.linspace(-5,5,500)
y = np.linspace(-5,5,500)

X, Y = np.meshgrid(x,y)
Z = (1-X/2+X**3+Y**4) * np.exp(-X**2-Y**2)

plt.figure(figsize=(8,8))
plt.contourf(X,Y,Z)
plt.show()
```
